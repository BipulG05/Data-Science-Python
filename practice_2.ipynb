# Missing value imputation by Mean , Median

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import BytesIO
```

```python
df=pd.read_csv('D:\\python programs\\CSV File\\train.csv')
```

```python
df
```

```python
df.info()
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.isnull().sum()
```

```python
null_val_per=df.isnull().sum()/df.shape[0]*100
null_val_per
```

```python
drop_columns_pre20=null_val_per[null_val_per > 20].keys()
drop_columns_pre20
```

```python
df2_drop_clm=df.drop(columns=drop_columns_pre20)
df2_drop_clm.shape
```

```python
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var=[ var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8,'color':'red'},label="original")
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5,'color':'green'},label="Mean")
    plt.legend()
    
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':9,'color':'red'},label="original")
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':6,'color':'green'},label="Mean")
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':2,'color':'k'},label="Medium")
    plt.legend()
    
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(16,9))
    plt.subplot(3,1,1)
    sns.boxplot(df3_num[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,2)
    sns.boxplot(df5_num_median[var])
```

```python
df_concat=pd.concat([df3_num[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]],axis=1)
```
